```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
```

```python
AUDIO_FILE = '/Users/anatrini/Documents/in_progress/artificial_models_HfMT/Block2/audio/whispers/whisper_01.wav'
samples, sample_rate = librosa.load(AUDIO_FILE, sr=None)
```

```python
plt.figure(figsize=(14,5))
librosa.display.waveshow(samples, sr=sample_rate)
```

```python
print('Whisper sample ', samples.shape, 'Sample rate ', sample_rate, 'Data type ', type(samples))
print(samples[22400:22420])
```

```python
# Display Spectrogram
sgram = librosa.stft(samples)
amp = np.abs(sgram)
librosa.display.specshow(amp)
```

```python
# Display spectrogram using mel scale instead of raw frequency
sgram_mag, _ = librosa.magphase(sgram)
mel_scale_sgram = librosa.feature.melspectrogram(S=sgram_mag, sr=sample_rate)
librosa.display.specshow(mel_scale_sgram)
```

```python
# Display spectrogram using decibel scale to get the final Mel Spectrogram
mel_sgram = librosa.amplitude_to_db(mel_scale_sgram, ref=np.min)
librosa.display.specshow(mel_sgram, sr=sample_rate, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
```
